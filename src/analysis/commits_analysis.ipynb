```python
results = {}

for student in d:
#     print("%s\n%s\n%d" % (student["name"], student["url"], student["ncommits"]))
#     print(json.dumps(student["commits"], indent=4))
#     print("\n")
    
    results.update({student["name"]: {}})
    
    for task_name in tasks:
        results[student["name"]].update({task_name: [0, 0, 0]})
        
    for commit in student["commits"]:
        
        if "github" in commit["url"]:            
            date_str = commit["date"].split("GMT")[0].strip()
            ddate = datetime.strptime(date_str, "%b %d, %Y, %I:%M %p")
        else:
            print("%s: %s" % (student["name"], commit["date"],))
            ddate = datetime.strptime(commit["date"], "%b %d, %Y %I:%M%p GMT%z")
            ddate = ddate.replace(tzinfo=None)
        
        commit_results = {}
        for task_name in tasks:
            commit_results.update({task_name: [0, 0, 0]})

        for file in commit["files"]:
            for task_name, task in tasks.items():
                if file and task_name in file:
                    l1 = task["time_slot_1"][0]
                    r1 = task["time_slot_1"][1]
                    
                    l2 = task["time_slot_2"][0]
                    r2 = task["time_slot_2"][1]
                    
                    if ddate >= l1 and ddate <= r1:
                        commit_results[task_name][0] = 1
                    elif ddate > l2 and ddate < r2:
                        commit_results[task_name][1] = 1
                    else:
                        commit_results[task_name][2] = 1
    
        for task_name in tasks:
            results[student["name"]][task_name][0] += commit_results[task_name][0]
            results[student["name"]][task_name][1] += commit_results[task_name][1]
            results[student["name"]][task_name][2] += commit_results[task_name][2]
            
    for task_name in tasks:
        results[student["name"]][task_name] = str(results[student["name"]][task_name])
    
print("Results:")
print(json.dumps(results, indent=4, ensure_ascii=False))
    
```
